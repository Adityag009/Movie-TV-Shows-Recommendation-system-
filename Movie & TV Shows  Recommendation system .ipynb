```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
# Stemmers remove morphological affixes from words, leaving only the word stem.
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

# Textvectorization
#### Text Vectorization is the process of converting text into numerical representation.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=15000,stop_words='english')
```

#### The CountVectorizer is a method from the scikit-learn library used for converting a collection of text documents into a matrix of token counts

##### The CountVectorizer is initialized with the desired parameters. In this case, max_features is set to 15000, which means that the maximum number of features (i.e., unique words) to be extracted from the text will be limited to 15000. stop_words is set to 'english', which means that common English words such as 'the', 'and', 'in', etc. will be excluded from the text.

##### CountVectorizer is a method for vectorizing text data by converting it into a matrix of token counts, which can be used for text classification or other natural language processing tasks.

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

##### Once the CountVectorizer is initialized, the fit_transform() method is called on a list of text documents. This method fits the model to the data and transforms the data into a matrix of token counts. Each row of the matrix represents a document, and each column represents a unique word in the corpus. The value in each cell of the matrix represents the number of times that word appears in that document.

```python
from sklearn.metrics.pairwise import cosine_similarity
similartity = cosine_similarity(vectors)
```

##### cosine_similarity is a function from the scikit-learn library used for computing the cosine similarity between pairs of vectors. The function takes a matrix of vectors as input and returns a similarity matrix where each element represents the cosine similarity between two vectors.

##### Cosine similarity is a measure of similarity between two non-zero vectors of an inner product space. It is defined as the cosine of the angle between the two vectors.

##### In the context of the cosine_similarity function from scikit-learn, the input matrix of vectors is usually a bag-of-words or TF-IDF matrix, and the output is a similarity matrix that can be used to find similar documents or clusters of documents.

```python
sorted(list(enumerate(similartity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    distances = similartity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title, ":",df.iloc[i[0]].type)
        
    return
```

```python
recommend('Sophie: A Murder in West Cork')
```
